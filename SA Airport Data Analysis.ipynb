# Pre-processing Data



```python
def bi_entry_handling(list):
    Col1=list[0:len(list):2]
    Col2=list[1:len(list):2]
    return Col1, Col2
```

```python
def tri_entry_handling(df):
    trip=[]
    flight_number=[]
    for all in df:
        if all =="Direct":
            trip.append(all)
        elif all == "Codeshare":
            trip.append(all)
        else:
            flight_number.append(all)
    return trip, flight_number
```

## Clean Missing Data

```python
def handle_missing_values(df):
    indicator = 0

    for all in df:
        indicator = 0
        na = df[all].isna().sum()
        null = df[all].isnull().sum()
        if na != 0:
            print(f"NaN values for {all}: {na}.\nZero values for {all}: {null}.")
            df.drop(columns=all)#, inplace =True)
            print(f"{all} rows removed")
            indicator += 1

        elif null != 0:
            print(f"NaN values for {all}: {na}.\nZero values for {all}: {null}.")
            df[all].replace(0, df[all].mean().sum())# ,inplace =True)
            print(f"{all} column filled with mean")
            indicator += 1
        
    if indicator == 0:
        print("No missing values found")

handle_missing_values(Unified)
Unified
```

# EDA

#### Which airline has the most flight records?

```python
airlines=Unified.groupby("Flight_Number")["Flight_Number"].unique().tolist()
df=Unified.groupby("Flight_Number")[["Flight_Number"]]
count=df.value_counts().to_list()
dff=Unified.groupby("Flight_Number")["Flight_Number"]
total_airlines = Unified.groupby("Flight_Number")["Flight_Number"].value_counts().sum()

total_airline_percentages=[]
for all in range(0,49):
    total_each_airline=(dff.value_counts())[all]
    pct=total_each_airline/total_airlines*100
    total_airline_percentages.append(f"{round(pct,2)}%")

labels = []
Count=0

relevant_data=pd.DataFrame(columns=["Airline","Count","Pct"])
lines=[]
for all in airlines:
    lines.append(all[0])
relevant_data["Airline"]=lines
relevant_data["Count"]=count
relevant_data["Pct"]=total_airline_percentages
sorted=relevant_data.sort_values(by="Count",ascending=False)

Count2=0
for all in sorted["Count"]:
    labels.append(f"{sorted["Airline"].values[Count2]} -  {sorted["Count"].values[Count2]} flight(s)  ({sorted["Pct"].values[Count2]})")
    Count2+=1

fig, ax = plt.subplots()

ax.pie(x=sorted["Count"], startangle=90, normalize=True)
ax.legend(labels,bbox_to_anchor=(1.1,2))
ax.set_title("Distribution of Flight Records for each Airline")
```

##### South African Airways (SA) has the most flight records as it's records compose 72.13% of all the flight records... followed by British Airways (BA) composing 3.18% and Mango (JE) composing 2.93% of the records


#### What location has the most flights recorded from?

```python
from_locations=Unified.groupby("From")["From"].unique().to_list()
from_count=Unified.groupby("From")["From"].value_counts().to_list()
total_from_locations = Unified.groupby("From")["From"].value_counts()

total_from_location_percentages=[]
limit=range(0,len(total_from_locations))
for all in limit:
    total_from_location=total_from_locations[all]
    pct=total_from_location/(total_from_locations.sum())*100
    total_from_location_percentages.append(f"{round(pct,2)}%")

from_=[]
count=0
limit=len(from_locations)+1
for all in from_locations:
   from_.append(all[0])
   count+=1
   if count==limit:break

df=pd.DataFrame(columns=["From","Count","Pct"])
df["From"]=from_
df["Count"]=from_count
df["Pct"]=total_from_location_percentages
dff=df.sort_values(by="Count",ascending=False)

labels = []
Count=0
for all in dff["From"]:
    labels.append(f"{dff["From"].values[Count]} - {dff["Count"].values[Count]} flight(s)  ({dff["Pct"].values[Count]})")
    Count+=1
    if Count==(len(dff["From"])+1):
        break

fig, ax = plt.subplots()
ax.pie(x=dff["Count"].to_list(), startangle=90, normalize=True)
ax.legend(labels,bbox_to_anchor=(1.1,2))
ax.set_title("Distribution of Flight Records From each Location")
```

##### JNB has the most flight records traveled from due to JNB composing 54.52% of the flight records, the runner ups CPT compose 13.45% and DUR compose 2.69% of the records. Since JNB has the most flights records we can get more accurate insights on JNB because we have more data on the location. It makes sense to draw more focus of the location JNB
